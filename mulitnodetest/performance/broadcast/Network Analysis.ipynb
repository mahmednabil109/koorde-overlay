```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import re, os
from pathlib import Path

log_files = list(Path('.').glob('log_out_?.log'))
broad_data = []

for i,f in enumerate(log_files):
    text = f.read_text()
    
    rts = re.findall(r"RT (\d+(\.\d+)?)", text)
    rts = list(map(lambda x: float(x[0]), rts))
    
    tps = re.findall(r"TPS (\d+(\.\d+)?)", text)
    tps = list(map(lambda x: float(x[0]), tps))
    broad_data.append([1 << (i+3), rts[0], tps[0]])

broad_data = pd.DataFrame(broad_data, columns=['Node Num', 'RT', 'TPS'])
broad_data
```

```python
plt.figure(figsize=(12,7))

sns.set(font_scale = 1.3)

sns.set_style("whitegrid")
p = sns.lineplot(x=broad_data['Node Num'], y=broad_data['RT'])
p.set(xlabel="Number of Nodes", ylabel="AVG Response Time")

plt.show()
```

```python
plt.figure(figsize=(12,7))

sns.set_style("whitegrid")
p = sns.lineplot(x=broad_data['Node Num'], y=broad_data['TPS'])
p.set(xlabel="Number of Nodes", ylabel="AVG Transaction Per Second")

plt.show()
```
