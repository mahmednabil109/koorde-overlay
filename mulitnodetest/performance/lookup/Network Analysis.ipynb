```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import re, os
from pathlib import Path

log_files = list(Path('.').glob('log_out_?.log'))
lookup_data = []

log_files = sorted(log_files, key=lambda a: a.name)

for i,f in enumerate(log_files):
    text = f.read_text()
    
    rts = re.findall(r"RT (\d+(\.\d+)?)", text)
    rts = list(map(lambda x: float(x[0]), rts))
    
    tps = re.findall(r"TPS (\d+(\.\d+)?)", text)
    tps = list(map(lambda x: float(x[0]), tps))
    lookup_data.append([1 << (i+3), rts[0], tps[0]])

lookup_df = pd.DataFrame(lookup_data, columns=['Node Num', 'RT', 'TPS'])
lookup_df
```

```python
plt.figure(figsize=(12,7))
sns.set(font_scale = 1.3)

sns.set_style("whitegrid")
p = sns.lineplot(x=lookup_df['Node Num'], y=lookup_df['RT'])
p.set(xlabel="Number of Nodes", ylabel="AVG Response Time")

plt.show()
```

```python
plt.figure(figsize=(12,7))

sns.set_style("whitegrid")
p = sns.lineplot(x=lookup_df['Node Num'], y=lookup_df['TPS'])
p.set(xlabel="Number of Nodes", ylabel="AVG Transaction Per Second")

plt.show()
```
